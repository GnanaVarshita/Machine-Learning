```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x= np.linspace(0,10,100)# Evenly spaces 100 values between 0 and 10
y= np.sin(x)
z = np.cos(x)
```

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

Plotting Sin and Cos

```python
#sin Wave
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,z)
plt.show()
```

```python
#adding title x-axis and y-axis
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine value')
plt.title('sine wave')
plt.show()
```

```python
#plotting a parabola

x= np.linspace(-10,10,20)
y= x**2 
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,'r+')
plt.show()
```

```python
plt.plot(x,y,'g.')
plt.show()
```

```python
plt.plot(x,y,'bx')
plt.show()

# refer matplotlib documentation
```

Plotting multiple graphs

```python
x= np.linspace(0,10,100)
plt.plot(x,np.sin(x),'r-')
plt.plot(x,np.cos(x),'g--')
```

Bar Plot

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
languages = ['Eng','French','Spanish','Latin','German']
ppl=[100,50,20,60,70]
ax.bar(languages,ppl)
plt.xlabel('Languages')
plt.ylabel('people')
plt.show()
```

Pie Chart

```python
fig1 = plt.figure()
ax= fig1.add_axes([0,0,1,1])
languages = ['Eng','French','Spanish','Latin','German']
ppl=[100,50,20,60,70]
ax.pie(ppl,labels=languages,autopct='%1.1f%%')
plt.show
```

```python
# 3D plot
fig3 = plt.figure()
ax = plt.axes(projection ='3d')
z = 20 * np.random.random(100)
x = np.sin(z)
y= np.cos(z)

ax.scatter(x,y,z,c=z,cmap ='Blues')
plt.show()
```
